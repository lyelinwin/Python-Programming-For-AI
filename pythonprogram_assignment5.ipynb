```python
import pandas as pd

# (1) How many rows do we have?
titanic = pd.read_csv('titanic.csv')
print("Number of rows:", len(titanic))
```

```python
# (2) How many columns do we have?
print("Number of columns:", len(titanic.columns))
```

```python
# (3) Which columns are numeric in value?
#titanic.info()
num = titanic.describe()

names = list(num.columns)
print("Columns that are numeric in value: ", names)
```

```python
# (4) How many people were traveling to New York City?
#.value_counts refers to how many times a particular value appears in data set

titanic["home.dest"].value_counts()
print("Number of people traveling to NYC: ",(titanic["home.dest"] == "New York, NY").sum())
```

```python
# (5) Which fare was bought the most?
titanic["fare"].value_counts()

#.mode() to get the most occurred 
most = titanic["fare"].mode()[0]
print("Fare bought the most:", most)
```

```python
# (6) What is the highest paid fare?
highest = titanic["fare"].max()
print("Highest paid fare: ", highest)
```

```python
# (7) How many females survived the crash?
# compound query to get sex & survivors for female survivors only

fem_num = titanic[(titanic["sex"] == "female") & (titanic["survived"] == 1)]
print("Number of females that survived the crash:", fem_num.shape[0])
```

```python
# (8) How many males survived the crash?
male_num = titanic[(titanic["sex"] == "male") & (titanic["survived"] == 1)]
print("Number of males that survived the crash: ", male_num.shape[0])
```

```python
# (9) How many survivors purchased a 2nd class ticket?
# pclass refers to passenger class (1st, 2nd, 3rd)

sec_tick = titanic.query('pclass == 2 and survived == 1')
print("Survivors with 2nd class ticket: ", sec_tick.shape[0])
```

```python
# (10) Draw a pie plot to show the age count for the survivors.
# get survivors with simple query
surv = titanic[titanic["survived"] == 1]

# filtering more in surv to get age 
age_survived = surv["age"].value_counts()
print(age_survived.plot(kind = "pie", figsize = (11,11), autopct = '%.1f%%'))
```

```python
# (11) Draw a bar plot of the count of people in each boat. Do not include the plot the '?' value.
# using simple query
no_quest = titanic[titanic["boat"] != "?"]
boat_count = no_quest["boat"].value_counts()

#this lets us know how many there are
len(no_quest["boat"].value_counts())

# bar plot
boat_count.head(27).plot(kind = "bar")
```
